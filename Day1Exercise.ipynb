```python
### import environments

#### local resources
import os
from IPython.display import Markdown, display
from openai import OpenAI

#### cloud resources
from dotenv import load_dotenv
from openai import OpenAI



```

```python
### Get the keys
# Load environment variables in a file called .env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")


```

```python
### connect to the models

#### OPen AI cloud

openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5", messages=[{"role": "user", "content": "Tell me a fun fact"}])

joke_openAI =  response.choices[0].message.content



#### Lamma local model

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
response = ollama.chat.completions.create(model="llama3.2", messages=[{"role": "user", "content": "Tell me a fun fact"}])

joke_lamma3 = response.choices[0].message.content



#### Deep seek local model

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
response = ollama.chat.completions.create(model="deepseek-r1:1.5b", messages=[{"role": "user", "content": "Tell me a fun fact"}])

joke_deepseek = response.choices[0].message.content

```

```python
### print beautiful results

display(Markdown(joke_openAI)) 
display(Markdown(joke_lamma3))
display(Markdown(joke_deepseek))


```

```python
jokes = {
    "openai": joke_openAI,
    "llama": joke_lamma3,
    "deepseek": joke_deepseek
}

winner = min(jokes, key=lambda k: len(jokes[k]))

print(winner, jokes[winner])
```
